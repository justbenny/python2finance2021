### Speed Comparison

```python
import random
I = 5000
```

```python
%time mat = [[random.gauss(0, 1) for j in range(I)] \
             for i in range(I)]  
```

```python
mat[0][:5]  
```

```python
%time sum([sum(l) for l in mat])  
```

```python
%time mat = np.random.standard_normal((I, I))  
```

```python
%time mat.sum()  
```

```python
r  
```

```python
s  
```

```python
r + s  
```

```python
2 * r + 3  
```
