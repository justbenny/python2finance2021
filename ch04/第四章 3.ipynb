```python
#if 的基本用法
flag = False
name = 'luren'
if name == 'python':       #判断变量是否为
    flag = True            #条件成立时标志为真
    print('welcome boss')  #输出欢迎信息
else:
    print(name)            #条件不成立时输出变量值
```

```python
fruits = ['banana','apple','mango']
for index in range(len(fruits)):
    print('当前序号：', index)
    print('当前水果：', fruits[index])
```

```python
for num in range(10,20):
    for i in range(2,num):
        if num%i == 0:
            j = num/i
            print('%d 等于 %d * %d'%(num,i,j))
            break
    else:
        print(num,'是一个质数')
```
